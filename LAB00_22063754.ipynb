```python
val = 'a'
val = 1

arr = []                        # empty array
arr = [1, 2, 3, 4, 5]          # numbers
arr = ['a','b','c','d','e']    # alphabets
arr = [1,'a',2,'b',3,'c']    # mixed types

# Zero-indexed: arr[0] = 1, arr[1] = 'a'
# Negative indexing: arr[-1] = 'c' (last element)
# Slicing: arr[start:end:step]
arr[1:5]     # ['a', 2, 'b', 3]
arr[1:5:2]   # ['a', 'b']
arr[:3]      # [1, 'a', 2]

# Modify arrays
arr.append(4)              # add one element to end
arr.extend(['d',5,'e'])   # add multiple elements
arr[0] = 0                 # assign value at index
arr.insert(1, 1)           # insert at index
del arr[0]                 # remove at index

joint_arr = list(zip([1,2,3],['a','b','c']))
# [(1,'a'),(2,'b'),(3,'c')]

# for loop
for a in arr:
    print(a)

# zip — loop two arrays at once
for item in zip(['a','b'],['artificial','breadth']):
    print(item[0] + ' for ' + item[1])

# enumerate — get index and value
for (index, item) in enumerate(['a','b','c','d']):
    print('Index of ' + item + ' is: ' + str(index))

# loop a dict
x_dict = {'d':'depth', 'e':'estimation', 'f':'frontier'}
for key, value in x_dict.items():
    print(key + ' for ' + value)

# while loop
x = 0
while x != 10:
    x += 1
    print(x)

# Basic function
def custom_fcn():
    print('This is a custom function')

# With required argument
def custom_fcn(msg):
    print('Display: ' + msg)

# With optional argument (default value)
def custom_fcn(msg='default message'):
    print('Display: ' + msg)

custom_fcn(msg='new message')  # named argument
```
